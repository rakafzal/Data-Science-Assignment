```python
numerical_features = df.select_dtypes(include=['int64', 'float64']).columns
```

```python
scaler_standard=StandardScaler()
```

```python
df[numerical_features] = scaler_standard.fit_transform(df[numerical_features])
```

```python
df
```

```python
one_hot_cols = [col for col in df.columns if df[col].nunique() < 5]
one_hot_encoder = OneHotEncoder()
one_hot_encoded = one_hot_encoder.fit_transform(df[one_hot_cols])
```

```python
label_cols = [col for col in df.columns if df[col].nunique() >= 5]
label_encoder = LabelEncoder()
label_encoded = df[label_cols].apply(label_encoder.fit_transform)
```

```python
encoded_df = pd.concat([pd.DataFrame(one_hot_encoded.toarray(), columns=one_hot_encoder.get_feature_names(one_hot_cols)),label_encoded], axis=1)
```

```python
df=pd.DataFrame(encoded_df)
```
