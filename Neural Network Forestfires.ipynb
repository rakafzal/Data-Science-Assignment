```python
df_1= df.iloc[:,2:30]
```

```python
stand_scaler = StandardScaler()
```

```python
stand_scaler.fit(df_1)
```

```python
df_norm=stand_scaler.transform(df_1)
df_norm
```

```python
pca=PCA(n_components=28)
```

```python
pca_value=pca.fit_transform(df_norm)
pca_value
```

```python
var = pca.explained_variance_ratio_
var
```

```python
var1 = np.cumsum(np.round(var,decimals = 4)*100)
var1
```

```python
finalDf = pd.concat([pd.DataFrame(pca_value[:,0:24],columns=['pc1','pc2','pc3','pc4','pc5','pc6','pc7',
 'pc8','pc9','pc10','pc11','pc12','pc13','pc14',
 'pc15','pc16','pc17','pc18','pc19','pc20','pc21',
 'pc22','pc23','pc24']),
 df[['size_category']]], axis = 1)
finalDf.size_category.replace(('large','small'),(1,0),inplace=True)
finalDf
```

```python
array = finalDf.values
X = array[:,0:24]
Y = array[:,24]
```

```python
model = Sequential()
model.add(Dense(12, input_dim=24, activation='relu'))
model.add(Dense(8,activation='relu'))
model.add(Dense(1,activation='sigmoid'))
model.summary()
```

```python
plot_model(model,show_shapes=True)
```

```python
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
history=model.fit(X, Y, validation_split=0.3, epochs=5, batch_size=10)
```

```python
y_pred= model.predict(X)
```

```python
scores = model.evaluate(X, Y)
print("%s: %.2f%%" % (model.metrics_names[1], scores[1]*100))
```

```python
plt.plot(range(5),history.history['accuracy'],label='Train')
plt.plot(range(5),history.history['val_accuracy'],label="Test")
plt.legend()
```

```python
plt.plot(range(5),history.history['loss'],label='Train')
plt.plot(range(5),history.history['val_loss'],label="Test")
plt.legend()
```
